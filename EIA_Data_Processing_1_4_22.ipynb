```python
import os
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#file to load
#load sales(consumption)
#load monthly total fuel 

sales_load = 'Resources/sales_revenue copy.csv'
# monthly_total_fuel = 'Resources/Monthly_Total_Fuel.csv'

#store the data as Pandas DataFrame
sales_load_df = pd.read_csv(sales_load)
# monthly_total_fuel_df = pd.read_csv(monthly_total_fuel ,header = [0])
```

```python
'''
merge year and month to a datetime data type
remove , to make numbers whole convert to int
remove . 
'''
```

```python
# consumption data per sector and per states since 1990
# data can be useful in analyzing GDP per sectors, and or States
sales_load_df
```

```python
#Converting '.' in cells to null values
sales_load_df = sales_load_df.replace('.', '', limit=0)
```

```python
sales_load_df
```

```python
#take out the , change to ()
for row in sales_load_df:
    for column in sales_load_df:
        sales_load_df.replace(',','', regex=True, inplace=True)
```

```python
sales_load_df
```

```python
sales_load_df = sales_load_df.replace(r'^\s*$', np.nan, regex=True)
```

```python
sales_load_df
```

```python
#Extracting State and Data Status column into a new df
str_col_df = sales_load_df[['State','Data Status']]
str_col_df
```

```python
#Converting string-numerical values to float64 datatype
num_sales_df = sales_load_df[['Year','Month', 'R/Revenue/Thousand Dollars','R/Sales/Megawatthours', 'R/Customers/Count', 'R/Price/Cents/kWh','C/Revenue/Thousand Dollars', 'C/Sales/Megawatthours','C/Customers/Count', 'C/Price/Cents/kWh', 'I/Revenue/Thousand Dollars','I/Sales/Megawatthours', 'I/Customers/Count', 'I/Price/Cents/kWh','T/Revenue/Thousand Dollars', 'T/Sales/Megawatthours','T/Customers/Count', 'T/Price/Cents/kWh', 'O/Revenue/Thousand Dollars','O/Sales/Megawatthours', 'O/Customers/Count', 'O/Price/Cents/kWh','T/Revenue/Thousand Dollars.1', 'T/Sales/Megawatthours.1','T/Customers/Count.1', 'TPrice/Cents/kWh']].astype(float)
```

```python
sales_load_df.dtypes
```

```python
#merge num_sales_df and str_col_df
clean_df = pd.concat([num_sales_df, str_col_df],axis=1)
clean_df.describe()
```

```python
#Initializing a new df for the combined Year and Month column.
date_df = pd.DataFrame()

#
date_df['Date'] = pd.to_datetime(clean_df[['Year', 'Month']].assign(DAY=28))
date_df
```

```python
#Merging new date into data
clean_df = pd.concat([date_df, clean_df],axis=1)
clean_df
```

```python
cleaner_df = clean_df.drop(['Year','Month'], 1)
```

```python
cleaner_df.dtypes
```

```python
cleaner_df.to_csv('Resources/Clean_Electric_Consumption_data.csv')
```
